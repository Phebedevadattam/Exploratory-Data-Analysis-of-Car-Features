**5.2: Removing the Outliers**

Removing outliers is a very good practise.if dont remove the outliers it may result to less accuracy of our model.
Here i have used the IQR score technique to remove the outliers as we can c in the above plots,the points outside the box are the outliers.

```
# calculating the IQR values for different columns
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```
# Removes all the outliers
df = df[~((df < (Q1-1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

##**Step 9**:Split the dataset into 80 and 20 ratio and build a machine learning model with Price as the target variable 

```
# Setting the target variable
X = df[['Popularity', 'Year', 'HP', 'Cylinders', 'MPG-H', 'MPG-C']].values
y = df['Price'].values
```

```
#scalling the features
from sklearn.preprocessing import StandardScaler
scaled_X = StandardScaler()
scaled_y = StandardScaler()
X = scaled_X.fit_transform(X)
y = scaled_y.fit_transform(y.reshape(-1,1))
```

```
# creating/splitting dataframe into training set and testing set 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=0)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

##**Step 10:**Try different algorithms and check their performance over metrics like R square, RMSE, MAE etc and document your findings 


###**Multiple Linear Regression**

```
#Fitting multiple linear regression to training set
from sklearn.linear_model import LinearRegression
lm = linear_model.LinearRegression()
model = lm.fit(X_train, y_train)
```

```
#Making Predictions
MLR_predictions = lm.predict(X_test)
plt.figure(figsize=(10,5))
plt.scatter(y_test,MLR_predictions)
plt.show()
```

In regression model, the most commonly known evaluation metrics include:

->**R-squared (R2)**- which is the proportion of variation in the outcome that is explained by the predictor variables. In multiple regression models, R2 corresponds to the squared correlation between the observed outcome values and the predicted values by the model. The Higher the R-squared, the better the model.

->**Root Mean Squared Error (RMSE)**- which measures the average error performed by the model in predicting the outcome for an observation. Mathematically, the RMSE is the square root of the mean squared error (MSE), which is the average squared difference between the observed actual outome values and the values predicted by the model. So, MSE = mean((observeds - predicteds)^2) and RMSE = sqrt(MSE). The lower the RMSE, the better the model.

->**Mean Absolute Error (MAE)**-like the RMSE, the MAE measures the prediction error. Mathematically, it is the average absolute difference between observed and predicted outcomes, MAE = mean(abs(observeds - predicteds)). MAE is less sensitive to outliers compared to RMSE.

```
#Evaluation metrics
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test,MLR_predictions))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test,MLR_predictions)))
print('R2 Score:', metrics.r2_score(y_test,MLR_predictions))
```

###**Random Forest Regression**

```
#Fitting Random Forest regression to training set
from sklearn.ensemble import RandomForestRegressor
regression=RandomForestRegressor(n_estimators=100,)
regression.fit(X_train, y_train)
```

```
#Making Predictions
RFR_predictions = regression.predict(X_test)
plt.figure(figsize=(10,5))
plt.scatter(y_test,RFR_predictions)
plt.show()
```

```
#Evaluation metrics
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test,RFR_predictions))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test,RFR_predictions)))
print('R2 Score:', metrics.r2_score(y_test,RFR_predictions))
```

###**Support Vector Regression**

```
#Fitting Support Vector regression to training set
from sklearn.svm import SVR 
regr=SVR()
regr.fit(X_train, y_train)
```

```
#Making Predictions
SVR_predictions = regr.predict(X_test)
plt.figure(figsize=(10,5))
plt.scatter(y_test,SVR_predictions)
plt.show()
```

```
#Evaluation metrics
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test,SVR_predictions))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test,SVR_predictions)))
print('R2 Score:', metrics.r2_score(y_test,SVR_predictions))
```

```
#Below the graph shows performance of different algorithms
MLR = round(lm.score(X_test, y_test), 2)*100
SVR=round(regr.score(X_test, y_test), 2)*100
RFR=round(regression.score(X_test, y_test), 2)*100
scores=[MLR,SVR,RFR]
names=["Multiple linear regression","Support vector regression","Random forest regression"]
plt.figure(figsize=(10,5))
plt.bar(names,scores)
plt.title("Scores of diffrent models implemented", weight='bold',fontsize=14)
plt.xlabel("Different models")
plt.ylabel("Scores in percentage")
plt.show()
```
